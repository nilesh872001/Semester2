 web crawler, also known as a spider or a bot, is a program that systematically browses the World Wide Web, typically for the purpose of indexing or gathering information.

```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
code=requests.get("https://www.amazon.in")
```

```python
plain=code.text
```

```python
soup=BeautifulSoup(plain,"html.parser")
```

```python
for link in soup.find_all('a'):
    print(link.get('href'))
```
